```python
import numpy as np
import matplotlib.pyplot as plt

stacking_accuracy = np.load('/content/gdrive/MyDrive/N-BaIoT/stacking_accuracy.npy')
stacking_report = np.load('/content/gdrive/MyDrive/N-BaIoT/stacking_report.npy')
stacking_cm = np.load('/content/gdrive/MyDrive/N-BaIoT/stacking_cm.npy')
stacking_roc = np.load('/content/gdrive/MyDrive/N-BaIoT/stacking_roc_auc.npy')

bagging_accuracy = np.load('/content/gdrive/MyDrive/N-BaIoT/bagging_accuracy.npy')
bagging_report = np.load('/content/gdrive/MyDrive/N-BaIoT/bagging_report.npy')
bagging_cm = np.load('/content/gdrive/MyDrive/N-BaIoT/bagging_cm.npy')
bagging_roc = np.load('/content/gdrive/MyDrive/N-BaIoT/bagging_roc_auc.npy')

central_accuracy = np.load('/content/gdrive/MyDrive/N-BaIoT/central_accuracy.npy')
central_report = np.load('/content/gdrive/MyDrive/N-BaIoT/central_report.npy')
central_cm = np.load('/content/gdrive/MyDrive/N-BaIoT/central_cm.npy')
central_roc = np.load('/content/gdrive/MyDrive/N-BaIoT/central_roc_auc.npy')
```

```python
models = ['Stacking', 'Bagging']
accuracies = [stacking_accuracy, bagging_accuracy]

plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['blue', 'green'])
plt.ylim(0, 100)
plt.xlabel('Model Type')
plt.ylabel('Accuracy (%)')
plt.title('Comparison of Stacking and Bagging Model Accuracies')
plt.show()
```

```python
models = ['Stacking', 'Bagging']

# Function to calculate metrics from a confusion matrix
def calculate_metrics(cm):
  TP = cm[0][0]
  FP = cm[0][1]
  FN = cm[1][0]
  TN = cm[1][1]
  accuracy = (TP + TN) / (TP + FP + FN + TN) * 100
  precision = TP / (TP + FP) * 100 if (TP + FP) > 0 else 0
  recall = TP / (TP + FN) * 100 if (TP + FN) > 0 else 0
  f1_score = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
  return accuracy, precision, recall, f1_score

# Calculate metrics for both models
stacking_metrics = calculate_metrics(stacking_cm)
bagging_metrics = calculate_metrics(bagging_cm)

# Set up the plot
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
x = np.arange(len(metrics))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, stacking_metrics, width, label='Stacking', color='blue')
rects2 = ax.bar(x + width/2, bagging_metrics, width, label='Bagging', color='green')

# Add labels, title, and legend
ax.set_ylabel('Percentage (%)')
ax.set_title('Comparison of Stacking and Bagging Model Metrics')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
plt.ylim(0, 100)

plt.show()
```

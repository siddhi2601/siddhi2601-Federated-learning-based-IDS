```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import glob

glob.glob('/content/gdrive/MyDrive/N-BaIoT/1')
master_df=[]
for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*gaf_combo*.csv'):
  print(csv)
  gc_df = pd.read_csv(csv)


for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*gaf_junk*.csv'):
  print(csv)
  gf_df = pd.read_csv(csv)


for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*gaf_scan*.csv'):
  print(csv)
  gs_df = pd.read_csv(csv)

for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*gaf_tcp*.csv'):
  print(csv)
  gt_df = pd.read_csv(csv)


for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*gaf_udp*.csv'):
  print(csv)
  gu_df = pd.read_csv(csv)


for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*mirai_ack*.csv'):
  print(csv)
  ma_df = pd.read_csv(csv)

for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*mirai_scan*.csv'):
  print(csv)
  ms_df = pd.read_csv(csv)

for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*mirai_syn*.csv'):
  print(csv)
  msy_df = pd.read_csv(csv)

for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*mirai_udp*.csv'):
  print(csv)
  mu_df = pd.read_csv(csv)

for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*mirai_udpplain*.csv'):
  print(csv)
  mud_df = pd.read_csv(csv)

for csv in glob.glob('/content/gdrive/MyDrive/N-BaIoT/1/*benign*.csv'):
  print(csv)
  bg_df = pd.read_csv(csv)

bg_df=bg_df.sample(frac=0.25,replace=False)
gc_df=gc_df.sample(frac=0.25,replace=False)
gf_df=gf_df.sample(frac=0.5,replace=False)
gs_df=gs_df.sample(frac=0.5,replace=False)
gt_df=gt_df.sample(frac=0.15,replace=False)
gu_df=gu_df.sample(frac=0.15,replace=False)
ma_df=ma_df.sample(frac=0.25,replace=False)
ms_df=ms_df.sample(frac=0.15,replace=False)
msy_df_d=msy_df.sample(frac=0.25,replace=False)
mu_df=mu_df.sample(frac=0.1,replace=False)
mud_df=mud_df.sample(frac=0.27,replace=False)

bg_df['type']='benign'
mu_df['type']='mirai_udp'
gc_df['type']='gafgyt_combo'
gf_df['type']='gafgyt_junk'
gs_df['type']='gafgyt_scan'
gt_df['type']='gafgyt_tcp'
gu_df['type']='gafgyt_udp'
ma_df['type']='mirai_ack'
ms_df['type']='mirai_scan'
msy_df['type']='mirai_syn'
mud_df['type']='mirai_udpplain'

df_1 = pd.concat([bg_df,mu_df,gc_df,gf_df,gs_df,gt_df,gu_df,ma_df,ms_df,msy_df,mud_df],
               axis=0, sort=False, ignore_index=True)
```

```python
len(df_1)
```

```python
df_1.groupby('type')['type'].count()
```

```python
#shuffling rows
sampler=np.random.permutation(len(df_1))
df_1=df_1.take(sampler)
df_1.head()
```

```python
df_1 = df_1.replace([np.inf, -np.inf], np.nan)
```

```python
df_1.dropna(inplace=True)
df_1.isna().sum().sum()
df_1.shape
```

```python
df_1.drop_duplicates(inplace=True)
df_1.shape
```

```python
df_1.info()
```

```python
df_1['type'] = df_1['type'].apply(lambda x: 0 if x.startswith("benign") else 1)
df_1['type'].value_counts()
df_1.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
le = LabelEncoder()
for column in df_1.select_dtypes(include=['object']).columns:
    df_1[column] = le.fit_transform(df_1[column])
```
